```python
import json
import librosa
import IPython.display as ipd
```

```python
# file = "data_reseg/data/eng/quantile_0.4.jsonl"
# file = "/work/hdd/bbjs/peng6/yodas/data_reseg/data/zho/quantile_0.3.jsonl"
file = "/work/hdd/bbjs/peng6/yodas/data_reseg/data/jpn/quantile_0.3.jsonl"
line_num = 300

with open(file, 'r') as f:
    cnt = 0
    for line in f:
        if cnt == line_num:
            sample = json.loads(line.strip())
            break
        cnt += 1

speech, sr = librosa.load(
    sample['wav_path'], sr=16000, offset=sample['start_time'], duration=sample['end_time'] - sample['start_time']
)

print("------ Text ------")
print(sample['text'])
print("------ Prev Text ------")
print(sample['prev_text'])

ipd.Audio(speech, rate=sr)
```
